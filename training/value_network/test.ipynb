```python
import time, torch
from data_manager import board_to_tensor
from model import SmallResNetValue as Model
import chess

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = Model().to(device)
model.eval()

b = chess.Board()  # start
# pojedyncze wywo≈Çanie
t0 = time.time()
for _ in range(10):
    x = board_to_tensor(b).to(device)
    with torch.no_grad():
        _ = model(x)
t1 = time.time()
print("avg single eval (10 runs):", (t1-t0)/10)

# batchowy test (B=32)
batch = torch.cat([board_to_tensor(b) for _ in range(32)], dim=0).to(device)
t0 = time.time()
with torch.no_grad():
    _ = model(batch)
t1 = time.time()
print("batch eval (B=32):", (t1-t0))
```
