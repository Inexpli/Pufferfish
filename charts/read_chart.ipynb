```python
import pandas as pd
import matplotlib.pyplot as plt

path = "../charts/cn2_bn2_rlrop.csv"

# Odczyt opisu z komentarza
with open(path, "r") as f:
    first_line = f.readline()
    description = first_line.strip("#").strip()

# Odczyt danych pomijajc komentarz
df = pd.read_csv(path, comment="#")

try:
    plt.plot(df["epoch"], df["loss"], marker="o")
except KeyError:
    plt.plot(df["epoch"], df["train_loss"], marker="o")

plt.ylim(0, 0.5)
plt.xlabel("Epoka")
plt.ylabel("Train Loss")
plt.title("Krzywa strat (Loss)")
plt.text(df["epoch"].max()/2, 0.6, description, ha="center", fontsize=15)
plt.show()

try:
    plt.plot(df["epoch"], df["val_loss"], marker="o")
    plt.ylim(0, 10)
    plt.xlabel("Epoka")
    plt.ylabel("Validation Loss")
    plt.title("Krzywa strat (Validation Loss)")
    plt.text(df["epoch"].max()/2, 12, description, ha="center", fontsize=15)
    plt.show()
except KeyError:
    print("Brak danych walidacyjnych (val_loss)")
```
