```python
import pandas as pd
import matplotlib.pyplot as plt

# Odczyt opisu z komentarza
with open("../charts/cn2_bn2.csv", "r") as f:
    first_line = f.readline()
    description = first_line.strip("#").strip()

# Odczyt danych pomijajÄ…c komentarz
df = pd.read_csv("../charts/cn2_bn2.csv", comment="#")

plt.plot(df["epoch"], df["loss"], marker="o")
plt.ylim(0, 0.5)
plt.xlabel("Epoka")
plt.ylabel("Loss")
plt.title("Krzywa strat (Loss)")
plt.text(df["epoch"].max()/2, 0.6, description, ha="center", fontsize=15)
plt.show()
```
